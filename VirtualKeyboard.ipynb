```python
import mediapipe as mp
import cv2 
import pydirectinput 
import cvzone
```

```python
from cvzone.HandTrackingModule import HandDetector
detector = HandDetector(detectionCon = 0.8)
```

```python
cap = cv2.VideoCapture(0)
cap.set(3,1200)
cap.set(4,720)
```

```python
def drawAll(img,buttonList):
    for button in buttonList:
        x, y = button.pos
        w, h = button.size
        cvzone.cornerRect(img,(button.pos[0], button.pos[1],button.size[0], button.size[1]),20,rt=0,colorC=(255,255,255))
        cv2.rectangle(img,button.pos,(x+w,y+h),(0,180,255),cv2.FILLED)
        cv2.putText(img,button.text, (x+20, y+65), cv2.FONT_HERSHEY_PLAIN,4,(255,255,255),4)
    return img
```

```python
class Button():
    def __init__(self,pos,text,size=[85,85]):
        self.pos=pos
        self.size=size
        self.text=text
```

```python
keys=[["Q","W","E","R","T","Y","U","I","O","P"],
      ["A","S","D","F","G","H","J","K","L",";"],
      ["Z","X","C","V","B","N","M",",",".","\t"]]
```

```python
buttonList = []
for i in range(len(keys)):
    for j, key in enumerate (keys[i]):
        buttonList.append(Button([80*j + 30, 100 *i+50], key, size=[65, 65]))
```

```python
import pydirectinput
from time import sleep
while True:
    res, img = cap.read()
    if not res or img is None:
        print("Failed to grab frame from webcam")
        continue  # Skip this loop and try again
    hands, img = detector.findHands(img)

    img = drawAll(img, buttonList)

    if hands:
        lmList = hands[0]["lmList"]  # Get landmark list

        for button in buttonList:
            x, y = button.pos
            w, h = button.size
            
            
            
            if x < lmList[8][0] < x + w and y < lmList[8][1] < y + h:
                cv2.rectangle(img, (x - 5, y - 5), (x + w + 5, y + h + 5), (175, 0, 175), cv2.FILLED)
                cv2.putText(img, button.text, (x + 20, y + 65),
                            cv2.FONT_HERSHEY_PLAIN, 4, (255, 255, 255), 4)
                l, _, _ = detector.findDistance(lmList[8][:2], lmList[12][:2])
                print(l)
                
                if l<50:
                    pydirectinput.press(button.text.lower())
                    cv2.rectangle(img,button.pos,(x+w,y+h),(8,0,255),cv2.FILLED)
                    cv2.putText(img, button.text, (x + 20, y + 65), cv2.FONT_HERSHEY_PLAIN, 4, (255, 255, 255), 4)

                    sleep(0.15)

    cv2.imshow("Image", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
