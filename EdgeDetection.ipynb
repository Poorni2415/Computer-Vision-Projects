```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Eiffel.jpeg')
```

```python
cv2.imshow("Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

plt.imshow(img)
plt.axis('off') 
plt.title("Image")
plt.show()
```

```python
img.shape
```

```python
img=cv2.resize(img,(320,320))
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
gray.shape
```

```python
cany_img=cv2.Canny(gray,159,200)
```

```python
plt.imshow(cany_img)
plt.show()
```

```python
import numpy as np
kern=np.ones((1,1),np.uint8)
eimg = cv2.erode(cany_img,kern,iterations=1)
plt.imshow(eimg)
plt.show()

kern=np.ones((4,4),np.uint8)
dimg=cv2.dilate(cany_img,kernel,iterations=1)
plt.imshow(dimg)
plt.show()
```
